```python
# Dependencies 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress
```

```python
# import data
mouse_metadata_path = "Resources/Mouse_metadata.csv"
study_results_path = "Resources/Study_results.csv"
```

```python
# Read in the data
mouse_data = pd.read_csv(mouse_metadata_path)
mouse_data
```

```python
results_data = pd.read_csv(study_results_path)
results_data
```

```python
# Combine the two datasets
all_data = pd.merge(results_data,mouse_data,on='Mouse ID')

# Display the data table for preview
all_data.head()
```

```python
# Checking the number of mice
all_data['Mouse ID'].value_counts()
```

```python
#better check for number of mice
all_data['Mouse ID'].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
grouped_all_data = all_data.groupby(['Mouse ID','Timepoint']).count()['Tumor Volume (mm3)']
duplicate = grouped_all_data.loc[grouped_all_data > 1]
duplicate
```

```python
# Get all the data for the duplicate mouse ID. 
duplicate_df = all_data.loc[all_data['Mouse ID'] == 'g989']
duplicate_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = all_data.loc[all_data['Mouse ID'] != 'g989']
clean_data
```

```python
# Checking the number of mice in the clean DataFrame.
clean_data['Mouse ID'].nunique()
```
